```python
# 使用plt.bar画出直方图
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
train_data=pd.read_csv("train.csv")
plt.figure(figsize=(13,5))
ax=plt.bar(train_data.columns,train_data.isnull().sum())
plt.xlabel("Columns Names")
plt.ylabel("Missing Number")



# Attach a text label above each bar in *rects*, displaying its height
def autolabel(rects):
    for rect in rects:
        height=rect.get_height()
        plt.annotate('{}'.format(height),
                    xy=(rect.get_x() + rect.get_width() / 2, height),
                    xytext=(0, 1),  # 1 points vertical offset
                    textcoords="offset points",
                    ha='center', va='bottom')
autolabel(ax)



plt.show()
```

```python
# 使用plt.plot画出折线图
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
train_data=pd.read_csv("train.csv")
plt.figure(figsize=(13,5))

xList=train_data.columns
yList=train_data.isnull().sum()
plt.plot(xList,yList)
plt.xlabel("Columns Names")
plt.ylabel("Missing Number")

#使用下面方法增加数据标签
for x, y in zip(xList, yList):
    plt.text(x, y+0.5, str(y), ha='center', va='bottom', fontsize=10.5)
    
plt.show()
```

```python
import numpy as np
import pandas as pd
import matplotlib as plt
movies=pd.read_csv('tmdb_5000_movies.csv')
movies_title_date=pd.DataFrame(movies,columns=['title','release_date'])
TJ=movies_title_date.dropna()
TJ.info()
```

```python
TJ["release_date"]=pd.to_datetime(TJ["release_date"])
# 转换格式由object变成datetime64[ns]，才有后面提取
TJ["year"]=TJ["release_date"].dt.year
TJ["quarter"]=TJ["release_date"].dt.quarter
TJ["month"]=TJ["release_date"].dt.month
TJ
```

```python
CDate=TJ.pivot_table(index='year',aggfunc='count')
CDate
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(18,8))
plt.plot(CDate.index.values,CDate["title"],color="red")
plt.bar(CDate.index.values,CDate["title"],color="green")
plt.xlabel("Year")
plt.ylabel("Movies Number")
plt.title("MoviesNumber by yearly")
plt.show()
```
