```python
# 计算Female存活率 返回行数：df.shape[1]  返回列数：df.shape[0]   或者：len(df)
F=train_data[train_data["Sex"]=="female"]
FS=F[F["Survived"]==1]
FSurv=FS.shape[0]/F.shape[0]
FSurv
```

```python
# 计算Male存活率 返回行数：df.shape[1]  返回列数：df.shape[0]   或者：len(df)
M=train_data[train_data["Sex"]=="male"]
MS=F[F["Survived"]==1]
MSurv=MS.shape[0]/M.shape[0]
MSurv
```
